```python
import pandas as pd

# Load the dataset
file_path = "Food_Recipe_with_reviews_final(with_rating).csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Display first few rows
df.head()
```

```python
#Check for Missing Values:
df.isnull().sum()

```

```python
df=df.copy() 
#Handle Missing Values:

#Fill missing values based on common values:
df["cuisine"] = df["cuisine"].fillna("Unknown")
df["course"] = df["course"].fillna("Other")
df["diet"] = df["diet"].fillna("Mixed")

#Fill missing numeric values with the median:
df["prep_time (in mins)"] = df["prep_time (in mins)"].fillna(df["prep_time (in mins)"].median())
df["cook_time (in mins)"] = df["cook_time (in mins)"].fillna(df["cook_time (in mins)"].median())

#Fill Missing Ingredient Values:
df["ingredients_name"] = df["ingredients_name"].fillna("Unknown Ingredient")
df["ingredients_quantity"] = df["ingredients_quantity"].fillna("Not Specified")

# Fill Missing Description
df["description"] = df["description"].fillna("No description available")

```

```python
df.isnull().sum()
```

```python
#Remove Duplicates:
df = df.drop_duplicates()

# Show duplicate rows (excluding the first occurrence)
df[df.duplicated()]
# Count total duplicate rows (excluding the first occurrence)
duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows: {duplicate_count}")

```

```python
print(df.columns)
```

```python
#Standardize Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("(", "").str.replace(")", "")
```

```python
print(df.columns)
missing_cols = [col for col in ["cuisine", "course", "diet"] if col not in df.columns]
if missing_cols:
    print(f"Missing columns: {missing_cols}")


#Encode Categorical Variables (One-Hot Encoding)

df = pd.get_dummies(df, columns=["cuisine", "course", "diet"])
df.head()
```
