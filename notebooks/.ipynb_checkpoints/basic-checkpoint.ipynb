## Data Analyse

```python
# Load dataset and analyze
import pandas as pd

# Load dataset
df = pd.read_csv("D:/College Work/Minor 2/Flavour-Match-/Dataset/filled_food_recipes_final.csv")

# Display first few rows
print(df.head())

# Check dataset info
print(df.info())

# Check summary statistics
print(df.describe())
```

## Check for missing values

```python
# Count missing values in each column
print(df.isnull().sum())

# Visualize missing values
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cmap='viridis', cbar=False, yticklabels=False)
plt.show()
```

## Check for duplicate values

```python
# Check for duplicate rows
print("Duplicate Rows:", df.duplicated().sum())

# Drop duplicates (if needed)
df = df.drop_duplicates()
```

## Explore Categorical data

```python
# Count unique cuisines
print(df["cuisine"].value_counts())

# Count unique diet types
print(df["diet"].value_counts())

# Visualize cuisine distribution
plt.figure(figsize=(10,10))
# Updated correct syntax
sns.countplot(data=df, 
              y="cuisine",
              hue="cuisine",  # Specify the hue parameter
              order=df["cuisine"].value_counts().index,
              palette="coolwarm",
              legend=False)  # Hide legend since it's redundant

plt.show()
plt.show()
```

## Analyse review and ratings

```python
# Check rating distribution
plt.figure(figsize=(8,4))
sns.histplot(df["rating"], bins=5, kde=True, color="blue")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.title("Distribution of Food Ratings")
plt.show()
```
